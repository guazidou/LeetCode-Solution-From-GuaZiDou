# Get API result

```python
import requests
```

```python
url = 'https://leetcode.com/api/problems/algorithms/'
```

```python
res = requests.get(url)
```

```python
res.status_code
```

```python
data = res.json()
```

```python
data
```

```python
problems = data['stat_status_pairs']
```

# Count open problems

```python
from collections import defaultdict
```

```python
stat = defaultdict(int)
```

```python
for each in problems:
    if each['paid_only']:
        stat['paid'] += 1
    else:
        stat['open'] += 1
```

```python
stat
```

# Serialize data

```python
import pickle
```

```python
with open('problems.plk', 'w') as f:
    pickle.dump(problems, f)
```
